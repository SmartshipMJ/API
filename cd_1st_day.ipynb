```python
score = 85

if score >= 90:
    print('A학점')      # score가 90 이상일 때
elif score >= 80:
    print('B학점')      # score가 80 이상일 때
elif score >= 70:
    print('C학점')      # score가 70 이상일 때
else:
    print('F학점')      # 그 외

# 출력: B학점
```

```python
# range()를 이용한 반복
for i in range(5):        # 0, 1, 2, 3, 4
    print(i)

# 리스트 순회
fruits = ['사과', '바나나', '포도']
for fruit in fruits:
    print(f'나는 {fruit}를 좋아합니다')

# range(시작, 끝, 간격)
for i in range(1, 10, 2):  # 1, 3, 5, 7, 9
    print(i)
```

```python
count = 0
while count < 5:
    print(f'count = {count}')
    count += 1  # count를 1씩 증가

# break: 반복문 즉시 종료
# continue: 이번 반복만 건너뛰고 다음 반복 진행
```

```python
for dan in range(2, 10):    # 2단 ~ 9단
    print(f'--- {dan}단 ---')
    for i in range(1, 10):  # 1 ~ 9
        print(f'{dan} x {i} = {dan * i}')
    print()  # 단 사이 빈 줄
```

```python
fruits = ['사과', '바나나', '포도', '딸기', '망고']

print(fruits[0])    # '사과'  (첫 번째)
print(fruits[2])    # '포도'  (세 번째)
print(fruits[-1])   # '망고'  (마지막)
print(fruits[-2])   # '딸기'  (뒤에서 두 번째)
print(fruits[::-1]) # 역순
print(fruits[1:3]) # 첫 번째부터 3번째 전 까지
print(fruits[:3])  # 3번째 전 까지
print(fruits[2:])  # 2번째부터
print(fruits[::2]) # 2칸씩
```

```python
nums = [3, 1, 4, 1, 5]
print('nums: ',nums)
nums.append(9)       # 끝에 추가: [3,1,4,1,5,9]
print('9 를 추가: ', nums)
nums.insert(0, 7)    # 특정 위치에 삽입: [7,3,1,4,1,5,9]
print('0번째에 7 삽입: ', nums)
nums.remove(1)       # 첫 번째 1 삭제: [7,3,4,1,5,9]
print('첫 번째 1 삭제: ', nums)
nums.pop()           # 마지막 요소 제거 후 반환: 9
print('pop: ', nums)
nums.sort()          # 정렬: [1,3,4,5,7]
print('정렬: ', nums)
nums.reverse()       # 역순: [7,5,4,3,1]
print('역순: ', nums)
print('길이:', len(nums))     # 길이: 5
```

```python
def draw_diamond(n):
    # 1. 상단 삼각형 (중앙 포함)
    for i in range(n):
        print(' ' * (n - i - 1) + '*' * (2 * i + 1))
    
    # 2. 하단 삼각형 (역삼각형)
    for i in range(n - 2, -1, -1):
        print(' ' * (n - i - 1) + '*' * (2 * i + 1))

# 높이가 n인 다이아몬드 출력
draw_diamond(10)
```
