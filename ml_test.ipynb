```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.metrics import mean_absolute_error
```

```python
df_normal = pd.read_csv('data/data/current_normal.csv')
df_anomaly = pd.read_csv('data/data/current_anomaly.csv')
```

```python
# 수치형 데이터만 필터링하고 결측치 있다면 제거
df_normal_numeric = df_normal.select_dtypes(include=[np.number])
df_normal_numeric = df_normal_numeric.dropna()
df_anomaly_numeric = df_anomaly.select_dtypes(include=[np.number])
df_anomaly_numeric = df_anomaly_numeric.dropna()
```

```python
# 모델 학습
model = IsolationForest(contamination=0.1, random_state=42)
model.fit(df_normal_numeric)
```

```python
# 예측 수행
preds_normal = model.predict(df_normal_numeric)
preds_test = model.predict(df_anomaly_numeric)
```

```python
# 성능 평가 (MAE)
score_test = model.decision_function(df_anomaly_numeric)
print(f"정상 데이터 내 감지된 이상치 개수: {list(preds_normal).count(-1)}")
print(f"테스트 데이터 내 감지된 이상치 개수: {list(preds_test).count(-1)}")
```

```python
# 이상 데이터 상세 확인
df_anomaly_numeric['anomaly_score'] = score_test
df_anomaly_numeric['is_anomaly'] = preds_test
print("\--- 이상치로 판정된 데이터 샘플 ---")
print(df_anomaly_numeric[df_anomaly_numeric['is_anomaly'] == -1].head())
```
